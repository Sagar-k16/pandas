#### .loc      ---> index location  --> used to extract elements based on USER DEFINED INDEX




#### .iloc     ---> index internal location  --> used to extract elements based on DEFAULT INDEX

+ .iloc
	+ objname.iloc[default_index_number]
	+ Extraction Rules are same 
        + upper bound exclusive
        + starts from 0 etc

```python
ser4
```

```python
ser4.iloc[0]
```

```python
ser4.iloc[1:5]
```

+ .loc
	+ objname.loc[user_index_number]
	+ Extraction Rules are different
		+ upper bound inclusive
		+ sarts from user defined index

```python
ser4.loc[11]
```

```python
ser4.loc[22:66]
```

## Data Frame

+ nothing but a collection of series (heterogeneous)

```python
dict_1={'col1':['a','b','c','d'], 'col2':[12,34,56,89],'col3':[12.45,67.43,78.55,67.33]}
dict_1
```

```python
pd.DataFrame(dict_1)
```

```python
column1=pd.Series(['a','b','c','d'],name='col1')
```

```python
column2=pd.Series([12,34,56,89],name='col2')
```

```python
column3=pd.Series([12.45,67.43,78.55,67.33],name='col3')
```

```python
table_ser=pd.concat([column1,column2,column3])   #it is in series format
table_ser
```

```python
table_ser=pd.concat([column1,column2,column3],axis=1)    #it is in table format
table_ser                                                #axis=1 for horizontal combine
                                                         #axis=0 fror vertically combine
```
